```python
import onnx
import onnx_graphsurgeon as gs

graph = gs.import_onnx(onnx.load("modernbert.onnx"))
```

```python
import graphviz
# Create a Graphviz object
dot = graphviz.Digraph(comment='BERT Model Visualization')
dot.attr(rankdir='TB')  # Top to bottom layout

# Add nodes
for node in graph.nodes:
    print(f"{str(id(node))}, {node.op}, {node.name}")
    dot.node(str(id(node)), f"{node.op}\\n{node.name}")

# Add edges
for node in graph.nodes:
    for input_name in node.inputs:
        for potential_source in graph.nodes:
            if input_name in potential_source.outputs:
                print(f"{str(id(potential_source))}, {str(id(node))}")
                dot.edge(str(id(potential_source)), str(id(node)))

# Render the graph
dot.render("bert_visualization", format="png", cleanup=True)
```
